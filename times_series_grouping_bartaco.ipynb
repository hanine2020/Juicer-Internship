```python
ts_list = []

errors_by_unit_dict = {}

df_features = pd.DataFrame()
i=0

for store in df.unit.unique():
    try:
        print(f'calculating features for UNIT {store}')
        aux = df[df['unit']==store]
        globals()[f'unit{store}_ts_volume'] = TimeSeriesData(aux[['hour','order_count']], time_col_name="hour")
        globals()[f'unit{store}_features'] = TsFeatures(selected_features =
                                                                    [  ####  reduce q of ts features calculated to reduce computing pw needed:
                                                                        # https://facebookresearch.github.io/Kats/api/_modules/kats/tsfeatures/tsfeatures.html
                                                                        'length',
                                                                        'mean','var',
                                                                        'entropy','lumpiness',
                                                                        'binarize_mean',
                                                                        'stability',
                                                                        'seasonality_strength',
                                                                        'level_shift_size',
                                                                        'seas_acf1',
                                                                        #'trend_num',v--> error: 'WARNING:root:Trend Detector failed "None of ['time'] are in the columns"
                                                                        'residual_std',
                                                                        'seasonal_period',
                                                                        #'hw_alpha', --> long to calc (x8)
                                                                        'trough',
                                                                        #'spikiness' --> too heavy to calc(x20)
                                                                    ]).transform( globals()[f'unit{store}_ts_volume'] )
        globals()[f'unit{store}_features']['store'] = store
        # print(type(globals()[f'{store[2:]}_features']))  --> dict
        # print(len(globals()[f'{store[2:]}_features']))  --> 40
        df_aux_feat = pd.DataFrame(globals()[f'unit{store}_features'], index=[i])
        
        df_features = pd.concat([df_features, df_aux_feat], axis=0)
        ts_list.append(f'unit{store}_ts_volume')
    except Exception as ex:
        # Generate and save tracebacks
        trace = []
        tb = ex.__traceback__
        while tb is not None:
            trace.append({
                "filename": tb.tb_frame.f_code.co_filename,
                "name": tb.tb_frame.f_code.co_name,
                "lineno": tb.tb_lineno
            })
            tb = tb.tb_next
        msg = str({
            'type': type(ex).__name__,
            'message': str(ex),
            'trace': trace
            })
        errors_by_unit_dict[store] = msg
        print(f'############## ERROR: Logged Error in unit {store} : {msg}')
    i+=1
print('##################  FINISHED  #####################')

print('')
print('ERRORS SAVED IN errors_by_unit_dict')
print(f'UNITS WITH ERROR: {errors_by_unit_dict.keys()}')
        
```

## what is X_train for tslearn?

https://tslearn.readthedocs.io/en/stable/gettingstarted.html#time-series-format

In tslearn, a time series is nothing more than a two-dimensional numpy array with its first dimension corresponding to the time axis and the second one being the feature dimensionality (1 by default).

Then, if we want to manipulate sets of time series, we can cast them to three-dimensional arrays, using to_time_series_dataset. If time series from the set are not equal-sized, NaN values are appended to the shorter ones and the shape of the resulting array is (n_ts, max_sz, d) where max_sz is the maximum of sizes for time series in the set.
